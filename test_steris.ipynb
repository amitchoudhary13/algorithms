```
from pyspark.sql.types import StructType,StructField, StringType, IntegerType,DateType
```

```
schema = StructType([ \
  StructField("PRODUCT",StringType(),True),\
  StructField("SALE_DATE",DateType(),True),\
  StructField("AMOUNT",IntegerType(),True)]
)
```

```

df = spark.read.csv("/FileStore/tables/test/sales.csv",header="true",schema=schema)
```

```
df.createTempView("sale")
```

```
spark.sql("select *,lag(sale_date,1) OVER (partition by product order by sale_date) as previous_date,lead(sale_date,1) OVER (partition by product order by sale_date) as Next_date from sale").show()
```
